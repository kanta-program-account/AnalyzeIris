```python
#最終課題
# k-means
# 特徴:クラスタ重心(クラスタセンタ)を見つけるアルゴリズム。
# メリット:クラスタが単純の形の場合上手く機能する。ベクトル量子化の考え方に基づき、クラスタセンタという各データポイントの重心を画像の成分(PCAやNMFの複数成分に相当)として使用できる為、PCAやNMFより強力な成分を見つけることが出来る。理解しやすく、高速。
# デメリット:重心だけでクラスタが定義される為、クラスタが複雑な形の場合は上手く機能しない。初期化が乱数で行なわれるため、再現性は比較的低い。探すクラスタの数をユーザーが指定しなければならない。

# dendrogram
# 特徴:データ点の近い順にクラスタリングされていくアルゴリズム。凝集型クラスタリングを階層に可視化したもの(階層型クラスタリング)。"近さ"の基準は ward, average, complete など複数種類ある。
# メリット:クラスタリングのプロセスを可視化できる。
# デメリット:2次元でしか使えない。クラスタが複雑な形の場合は上手く機能しない。

# DBSCAN
# 特徴:データポイントの密度に基づきクラスタリングを行うアルゴリズム。
# メリット:min_samples, epsパラメータにより、大小様々なデータセットにも適用可能。
# デメリット:処理が遅い。
```
